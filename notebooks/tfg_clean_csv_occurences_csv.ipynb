```python
import pandas as pd
import numpy as np

# RUTAS
input_parquet = r"../data/interim/embeddings/spain_covidMHstrict_occurrences_2020-03_2021-03_CLEANED.parquet"
output_csv = r"../data/interim/embeddings/spain_covidMHstrict_occurrences_2020-03_2021-03_CLEANED.csv"

print("📂 Cargando archivo limpio (Parquet)...")
df = pd.read_parquet(input_parquet)

# PREPARACIÓN PARA CSV (IMPORTANTE)
# Los arrays de numpy a veces dan problemas al guardarse en CSV. 
# Los convertimos a listas de Python estándar para que se guarden como texto plano "[0.1, 0.2, ...]"
print("🔄 Formateando arrays para CSV...")

emb_cols = [c for c in df.columns if 'embedding' in c]

for col in emb_cols:
    # Verificamos si es numpy array y lo pasamos a lista simple
    # Esto asegura que en el CSV se vea como: "[0.0123, 0.4321, ...]"
    df[col] = df[col].apply(lambda x: x.tolist() if isinstance(x, np.ndarray) else x)

print(f"💾 Guardando como CSV compatible en:\n   {output_csv}")
# index=False es vital para no crear una columna extra que rompa tu código
df.to_csv(output_csv, index=False) 

print("✅ ¡LISTO! Ahora tu código base leerá este archivo sin quejarse.")
```
